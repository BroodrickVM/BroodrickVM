Use the Washington DC publicly available crime data from 2009 to 2025 to identify two key insights that help to explain current trends for crime within the District. ## Identify and load all crime data files

### Subtask:
Find all the CSV files containing crime data and load them into a list of DataFrames.


**Reasoning**:
List files in the content directory, filter for relevant CSVs, load each into a DataFrame, and append to a list.



```python
import os

extracted_files = os.listdir('/content/')
crime_files = [f for f in extracted_files if 'Crime_Incidents_in_' in f and f.endswith('.csv')]

crime_dataframes = []
for file in crime_files:
    file_path = os.path.join('/content/', file)
    df = pd.read_csv(file_path, parse_dates=['REPORT_DAT'], index_col='REPORT_DAT')
    crime_dataframes.append(df)

print(f"Loaded {len(crime_dataframes)} dataframes.")
```

## Combine dataframes

### Subtask:
Concatenate all the yearly DataFrames into a single DataFrame for comprehensive analysis.


**Reasoning**:
Concatenate the list of DataFrames into a single DataFrame and display its head and shape.



```python
all_crime_data = pd.concat(crime_dataframes)
display(all_crime_data.head())
print(all_crime_data.shape)
```

## Data cleaning and preparation

### Subtask:
Handle missing values and ensure data types are appropriate for analysis and visualization.


**Reasoning**:
Display the percentage of missing values for each column to understand the extent of missing data.



```python
missing_percentages = all_crime_data.isnull().sum() / len(all_crime_data) * 100
display(missing_percentages)
```

**Reasoning**:
Drop columns with a high percentage of missing values and handle missing values in other relevant columns by dropping rows with missing values, then verify the data types.



```python
columns_to_drop = ['BID', 'OCTO_RECORD_ID']
all_crime_data.drop(columns=columns_to_drop, inplace=True)

all_crime_data.dropna(subset=['WARD', 'ANC', 'DISTRICT', 'PSA', 'NEIGHBORHOOD_CLUSTER', 'BLOCK_GROUP', 'CENSUS_TRACT', 'VOTING_PRECINCT', 'START_DATE', 'END_DATE'], inplace=True)

all_crime_data['START_DATE'] = pd.to_datetime(all_crime_data['START_DATE'], errors='coerce', utc=True)
all_crime_data['END_DATE'] = pd.to_datetime(all_crime_data['END_DATE'], errors='coerce', utc=True)

all_crime_data.dropna(subset=['START_DATE', 'END_DATE'], inplace=True)

display(all_crime_data.isnull().sum())
display(all_crime_data.info())
```

**Reasoning**:
The remaining missing values in the 'BLOCK' column are minimal, so dropping these rows will not significantly impact the dataset. After this, all missing values will be handled, and the data types are appropriate for analysis, completing the subtask.



```python
all_crime_data.dropna(subset=['BLOCK'], inplace=True)
display(all_crime_data.isnull().sum())
```

## Identify key insights

### Subtask:
Analyze the data to find two interesting trends or patterns in crime data.


**Reasoning**:
To find trends or patterns in crime data over time, I will group the data by year and count the number of incidents.



```python
crime_counts_by_year = all_crime_data.resample('Y').size()
display(crime_counts_by_year)
```

**Reasoning**:
To analyze the distribution of crime incidents by day of the week, I will extract the day of the week from the index and count the occurrences of each day.



```python
crime_counts_by_day = all_crime_data.index.day_name().value_counts()
display(crime_counts_by_day)
```

**Reasoning**:
To explore the spatial distribution of crime incidents, I will count the number of incidents in each ward.



```python
crime_counts_by_ward = all_crime_data['WARD'].value_counts().sort_index()
display(crime_counts_by_ward)
```

## Visualize insights

### Subtask:
Create two different visualizations (using Matplotlib or Seaborn) to demonstrate the identified insights: crime trends over the years and crime distribution by ward.


**Reasoning**:
Create a line plot for crime trends over the years and a bar plot for crime distribution by ward using the previously calculated counts, then display both plots.



```python
plt.figure(figsize=(12, 6))
plt.plot(crime_counts_by_year.index.year, crime_counts_by_year.values)
plt.title('Crime Incidents Over the Years (2009-2025)')
plt.xlabel('Year')
plt.ylabel('Number of Crime Incidents')
plt.grid(True)

plt.figure(figsize=(10, 6))
sns.barplot(x=crime_counts_by_ward.index, y=crime_counts_by_ward.values)
plt.title('Crime Incidents by Ward')
plt.xlabel('Ward')
plt.ylabel('Number of Crime Incidents')

plt.show()
```

## Refine visualizations

### Subtask:
Enhance the visualizations for clarity and impact, adding titles, labels, and legends as needed.


**Reasoning**:
Improve the visualizations by adding titles, labels, and adjusting figure sizes for clarity. For the line plot, ensure x-axis labels are readable. For the bar plot, add a descriptive title and axis labels. Adjust figure sizes for both plots for better display.



```python
plt.figure(figsize=(14, 7))
plt.plot(crime_counts_by_year.index.year, crime_counts_by_year.values)
plt.title('Annual Crime Incident Trends in Washington D.C. (2009-2025)', fontsize=16)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Number of Crime Incidents', fontsize=12)
plt.xticks(crime_counts_by_year.index.year, rotation=45, ha='right') # Rotate labels for readability
plt.grid(True)
plt.tight_layout() # Adjust layout to prevent labels overlapping

plt.figure(figsize=(12, 7))
sns.barplot(x=crime_counts_by_ward.index, y=crime_counts_by_ward.values, palette='viridis')
plt.title('Total Crime Incidents by Ward in Washington D.C. (2009-2025)', fontsize=16)
plt.xlabel('Ward', fontsize=12)
plt.ylabel('Number of Crime Incidents', fontsize=12)
plt.xticks(rotation=0) # Ensure ward labels are not rotated
plt.tight_layout() # Adjust layout

plt.show()
```
